```python
#TASK 1
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv("district wise population and centroids.csv")
tamil_nadu_df=df[df['State']=='Tamil Nadu']
plt.figure(figsize=(14, 8))
plt.bar(tamil_nadu_df['District'], tamil_nadu_df['Population in 2011'],color='blue')
plt.xlabel('District')
plt.ylabel('Population in 2011')
plt.title('Population in the Districts of Tamil Nadu')
plt.xticks(rotation=45)
plt.show()
```
